```python
import pickle
import matplotlib.pyplot as plt 
```

```python
def unpickle(file):
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
files = [f"datasets/cifar-10-batches-py/data_batch_{i}" for i in range(1,6)]

db1 = unpickle(files[0])
db1.keys()
```

```python
db1[b'data'][0]
```

# Visualize

```python
img = db1[b'data'][0]
# Example transformations to display an image
img = img.reshape(3, 32, 32)
img = img.transpose(1, 2, 0)

plt.imshow(img)
```

# Reformat All Data

```python
data = db1[b'data']
data = data.reshape(len(data), 3, 32, 32).transpose(0, 2, 3, 1)
data.shape
```

```python
rows = 4
columns = 4

fig = plt.figure(figsize=(8, 8))

for i in range(1, rows*columns+1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(data[i])
```
